```python
import requests
import json
import pandas as pd
from pandas.io.json import json_normalize
```

### Calling Discovery API

```python
params = {'apikey':'4XrkbIMbr4gS3vG1qsFiZpDcXSbyAZni', 'countryCode':'US', 'sort':'date,asc', 'size':'200'}

ticketmaster = requests.get('https://app.ticketmaster.com/discovery/v2/events.json', params=params)

ticketmaster_json = ticketmaster.json()
```

### Reading as DF

```python
first_df = json_normalize(ticketmaster_json)
first_df
```

```python
second_df = json_normalize(first_df['_embedded.events'][0])
second_df.dropna(axis=0, subset=[['_embedded.attractions', '_embedded.venues', 'classifications', 'priceRanges']], inplace=True)
second_df.head(15)
```

```python
second_df.columns
```

### Exploring _embedded.attractions

```python
attractions = pd.DataFrame()

for i in second_df.index:
    observation = json_normalize(second_df.loc[i, '_embedded.attractions'])
    attractions = attractions.append(observation)

# Potentially valuable features
attractions[['id','name','type','upcomingEvents._total']]
```

```python
attractions.columns
```

### Exploring _embedded.venues

```python
venues = pd.DataFrame()

for i in second_df.index:
    observation = json_normalize(second_df.loc[i, '_embedded.venues'])
    venues = venues.append(observation)

# Potentially valuable features
venues[['id','name','type','country.countryCode','state.name','city.name','upcomingEvents._total','upcomingEvents.ticketmaster']]
```

```python
venues.columns
```

### Exploring classifications

```python
classifications = pd.DataFrame()

for i in second_df.index:
    observation = json_normalize(second_df.loc[i, 'classifications'])
    classifications = classifications.append(observation)

# Potentially valuable features
classifications
```
